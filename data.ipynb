1 - Importar dados

```python
import pandas as pd
import plotly.express as px
import openpyxl as xl

data = pd.read_excel("new_clients.xlsx")

display(data)
```

2 - Limpar os Dados

```python
data = data.rename(columns={"join_date": "Entrada", "gender_id": "Genero"}).drop(["customer_id", "city_id"], axis = 1)

display(data)
```

```python
data["Genero"] = data["Genero"].replace({1:"Feminino",2:"Masculino"})

display(data)
```

3 - Checar Informações dos Dados

```python
data.info()
```

4 - Estruturar o Gráfico

```python
#Quant. de Entradas
entry = data.groupby('Entrada').size().reset_index(name='Data')


display(data)
```

```python
#Média gráfica + Criação do Gráfico
custom = entry.copy()
custom['Suave'] = custom['Data'].rolling(window=100).mean()
custom = custom.dropna()


fig = px.line(
    custom,
    x='Entrada',
    y='Suave',
    title= 'Média Entrada de Clientes',
    labels={'Entrada':'Data','Suave':'Média de entradas'},
    render_mode='svg'
)

fig.update_traces(
    line=dict(width=3, color='#FF0000'),
    mode='lines'
)

fig.update_layout(
    template='plotly_white',
    hovermode='x unified'
)

fig.show()
```
